```
import numpy as np
import matplotlib.pyplot as plt

def plotwave(x,y):
    plt.plot(x,y)
    plt.show()
    return None

x=np.pi*np.arange(200)/100
y=np.sin(x)

plotwave(x,y)
```

# 10 ファイルの処理

##10.1 ファイルに書き込み

```
filename='data.txt'  #ファイル名を'data,txt'と指定する。
f=open(filename,'w')  #ファイルを書き込みタイプ'w'として開く、fというテキスト処理型変数に渡す
print(type(f))　　#fはどいうタイプを調べる
f.write('Hello World') #ファイルに書き込む
f.close() #開いたファイルを閉じる
```

```
filename='data.txt'  #ファイル名を'data,txt'と指定する。
with open(filename,'w') as f:  #withのブロックにおいてfが書き込むことができる
    f.write('Hello World') #ファイルに書き込む
#自動的に閉じる
```

```
def getsin(n=100): # サイン波形を発生する関数
    x=np.pi*np.arange(2*n)/n
    y=np.sin(x)
    return x,y
def writewave(x,y,fn):          #ファイルに保存する関数
    with open(fn,'w') as f:     #withの範囲でファイルを開く
        f.write('%s %s\n' % ('x', 'y'))    # 軸名を書き込む、\nで改行
        for xi,yi in zip(x,y):                #xi,yiごとを使い、全ての要素を順番に使われる
            f.write('%f %f\n' % (xi,yi))     #６桁のうち小数点以下２桁
    return None
x,y=getsin()            #波形発生
writewave(x,y,'sin.txt')  #ファイルに書き込む
```

## 10.2 読み取り

```
filename='sin.txt'
with open(filename,'r') as f:
    contents=f.read()
lines=contents.split('\n')
X=[]
Y=[]
for line in lines[1:]:
    data=line.split()
    if len(data)==2:
        x=float(data[0])
        y=float(data[1])
        X.append(x)
        Y.append(y)
plt.plot(X,Y)
plt.show()
```
